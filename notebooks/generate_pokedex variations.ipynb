```python
import pandas as pd
import nlpaug
import nlpaug.augmenter.char as nac
import nlpaug.augmenter.sentence as nas
import nlpaug.augmenter.word as naw
from tqdm import tqdm
```

```python
pokedex = pd.read_pickle('poke_df.pkl')
```

```python
pokedex
```

```python
extended_descriptions = []
syn_aug = naw.SynonymAug(aug_p=0.15)
#sentence = list(pokedex.Description)[0]
for sentence in tqdm(list(pokedex.Description)):
    
    mod_sentence = syn_aug.augment(sentence, n=10)
    extended_descriptions.append(mod_sentence)
```

```python
extended_descriptions
```

```python
pokedex['desc_list'] = extended_descriptions
```

```python
pokedex
```

```python
big_df = pd.concat([pokedex.reset_index(),
          pd.DataFrame([pd.Series(x) for x in pokedex.desc_list]).reset_index()], axis = 1)
big_df
```

```python
big_df = big_df.drop(columns = ['index', 'desc_list'])
```

```python
big_df.to_csv('big_df.csv')
```

```python
pd.melt(big_df, id_vars = ['Pokemon']).to_csv('big_df.csv')
```

Let's do the same thing as above, but for the image captions that we generated the other day!

```python
descriptions = pd.read_csv('pokemon_image_descriptions.csv').drop(columns = ['Unnamed: 0'])
```

```python
descriptions
```

```python
extended_image_to_text = []
syn_aug = naw.SynonymAug(aug_p=0.15)
for sentence in tqdm(list(descriptions.Caption)):
    
    mod_sentence = syn_aug.augment(sentence, n=10)
    extended_image_to_text.append(mod_sentence)
```

```python
descriptions['desc_list'] = extended_image_to_text

big_df_images = pd.concat([descriptions.reset_index(),
          pd.DataFrame([pd.Series(x) for x in descriptions.desc_list]).reset_index()], axis = 1)
big_df_images = big_df_images.drop(columns = ['index', 'desc_list'])
big_df_images_melted = pd.melt(big_df_images, id_vars = ['Pokemon'])#.to_csv('big_df.csv')
```

```python
big_df_images_melted.to_csv('big_df_image_descriptions.csv')
```
