```python
import pandas as pd
import numpy as np
from bing_image_downloader import downloader
import os
from os import path
import shutil
from pathlib import Path
```

```python
poke_df = pd.read_pickle('poke_df.pkl')
```

```python
poke_df
```

 Use `bing_image_downloader` to get images of all the pokemon

```python
for pokemon in poke_df.Pokemon.unique():
    downloader.download(pokemon + str(' pokemon transparent background'),
    limit= 1,
    output_dir='images',
    adult_filter_off=True,
    force_replace=False,
    timeout=10,
    verbose=True)
```

See if there are any differences...

```python

set.difference(
    set([x.split(' pokemon transparent background')[0] for x in os.listdir('images')]),
    set(poke_df.Pokemon.unique())
)
```

No need to worry, that's just a hidden Mac File

Now, let's rename them all to the correct name.

```python
folder = "images"
for count, filename in enumerate(os.listdir(folder)):
    folder = f"images/{filename}"

    # rename the png's
    src = f"{folder}/Image_1.png"
    if path.exists(src):
        os.rename(src,f"{folder}/{filename.split(' pokemon transparent background')[0]}.png")

    # rename the jpg's
    src = f"{folder}/Image_1.jpg"
    if path.exists(src):
        os.rename(src,f"{folder}/{filename.split(' pokemon transparent background')[0]}.jpg")
```

`bing_image_downloader` creates a new directory for each of the pokemon within our images directory. Let's collapse those all down.

```python

directory = "images/"
for dirpath, _, filenames in os.walk(directory, topdown=True):
    
    source = os.path.join(dirpath, filenames[0])
    target = os.path.join(directory)


    # having some issues with a single pokemon, so I'll manually move the image once all these work
    if ('Azumarill' not in source) and ('.DS_Store' not in source):
        shutil.move(source,target)



# delete the empty folders
for file in os.listdir(directory):
    if path.isdir(f"images/{file}"):
        shutil.rmtree(f"images/{file}")
```
